```python
inicio = datetime.date(2020, 2, 25)
fin = datetime.date(2019, 3, 25)
months = -1
```

```python
fecha_final = add_months(inicio, -2)
contador = 1


trimestre = []
campeones = []
loyals = []
potential_loyal = []
news = []
promising = []
cant_lost = []
risk = []
need_atention = []
close_hibernating = []
hibernating = []

while fecha_final >= fin :
   
    fi_int = datetime_to_integer(inicio)
    ff_int = datetime_to_integer(fecha_final)
    print(f'Ciclo--{contador}:')
    print(f'La fecha de inicio es: {inicio}. La fecha final es: {fecha_final}.' )
    fecha_out = str(ff_int) # Fecha lmite inferior para el trimestre de inters
    fecha_in = str(fi_int)  # Fecha lmite superior para el trimestre de inters
    label = 'TOP'
    df = pd.read_parquet(f'TOP-{fecha_in}--{fecha_out}.parquet')
    print('=================================================================================================')
    
    rfm_model = rfm_scoring(df, fecha_in, fecha_out, label)
    print('-------------------------------------------------------------------------------------------------')

    #rfm_corr = rfm_model[['RECENCIA', 'FRECUENCIA', 'MONTO']].corr()
    #print('La matriz de correlacion es:')
    #print(rfm_corr.corr())
    #print('-------------------------------------------------------------------------------------------------')
    
    #bins_R = 90
    #bins_F = 100
    #bins_M = 100
    #range_R = [0,90]
    #range_F = [0, 10000]
    #range_M = [0, 10000000]
    #histogramas(rfm_model['RECENCIA'], rfm_model['FRECUENCIA'], rfm_model['MONTO'], 
    #        bins_R, bins_F, bins_M, range_R, range_F, range_M)

    #EDA...
    #histogramas(rfm['RECENCIA'], rfm['FRECUENCIA'], rfm['MONTO'])
    #print('-------------------------------------------------------------------------------------------------')
    
    segmentar(rfm_model, fecha_in, fecha_out)
    rfm_segmentos, dic_cuit, rfm_segmentado = label_segmentos(rfm_model, fecha_in, fecha_out,label)
    
     #Creando listas para generar grfico de lineas como funcin de los trimestres')
   
    entrada = datetime.datetime.strptime(f'{inicio}', '%Y-%m-%d').strftime('%d/%m/%Y')
    salida = datetime.datetime.strptime(f'{fecha_final}', '%Y-%m-%d').strftime('%d/%m/%Y')
   
    trimestre.append(f'{entrada}--{salida}')
    campeones.append(rfm_segmentado[rfm_segmentado['Segment'] == 'Campeones'][['%']].iloc[0,0])
    loyals.append(rfm_segmentado[rfm_segmentado['Segment'] == 'Leales'][['%']].iloc[0,0])
    potential_loyal.append(rfm_segmentado[rfm_segmentado['Segment'] == 'Potencialmente Leales'][['%']].iloc[0,0])
    news.append(rfm_segmentado[rfm_segmentado['Segment'] == 'Reciente operativo'][['%']].iloc[0,0])
    promising.append(rfm_segmentado[rfm_segmentado['Segment'] == 'Prometedores'][['%']].iloc[0,0])
    cant_lost.append(rfm_segmentado[rfm_segmentado['Segment'] == 'No se pueden perder'][['%']].iloc[0,0])
    risk.append(rfm_segmentado[rfm_segmentado['Segment'] == 'En Riesgo'][['%']].iloc[0,0])
    need_atention.append(rfm_segmentado[rfm_segmentado['Segment'] == 'Necesitan Atencion'][['%']].iloc[0,0])
    close_hibernating.append(rfm_segmentado[rfm_segmentado['Segment'] == 'Cercanos a Hibernar'][['%']].iloc[0,0])
    hibernating.append(rfm_segmentado[rfm_segmentado['Segment'] == 'Hibernando'][['%']].iloc[0,0])
      
   
    inicio = add_months(inicio, months)
    fecha_final = add_months(fecha_final, months)
    contador += 1
   
    print('=================================================================================================')
    
print('Evolucin trimestral de la poblacin en cada segmento...')   

dic = {'Trimestre': trimestre ,'Campeones': campeones , 'Leales': loyals, 'Potencialmente Leales': potential_loyal,
       'Reciente operativo': news, 'Prometedores': promising, 'No se pueden perder': cant_lost,'En Riesgo': risk ,
       'Necesitan Atencion': need_atention , 'Cercanos a Hibernar': close_hibernating , 'Hibernando': hibernating} 

df_lineas = pd.DataFrame(dic)
df_lineas.to_csv('1anio_porcentajes.csv', index = False)
# reverse the dataframe
df_lineas = df_lineas[::-1]
evolucion_trimestral_segmentos(df_lineas)


```
